```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.rcParams["font.family"] = "malgun gothic"
df = pd.read_csv("/Users/bugil/Downloads/primier_league.csv",encoding="utf-8")

df
```

```python
# 데이터 전처리 널 값이 존재하지 않는 것을 확인 할 수 있다.
df.isnull().sum()
```

```python
#EPL 프리미어 리그 득점 평균
goal_info = df['득점'].describe()
goal_info
```

```python
# EPL 프리미어리그 득점 순위 Top5
data_grp = df.groupby(['선수명']).sum()
data_srt = data_grp.sort_values(by='득점', ascending=False).head(5)
data_srt.drop(['경기수', '도움', '공격포인트', '슈팅', '유효슈팅'], axis=1, inplace=True)

data_srt
```

```python
# EPL 프리미어 리그 득점 순위 Top5 막대 그래프
label = ['케인', '살라', '페르난데스', '뱀포드', '손흥민']

index = np.arange((len(label)))

plt.rcParams["font.family"] = 'AppleGothic'
plt.bar(index, data_srt['득점'])
plt.title('EPL 득점 순위 Top5', fontsize = 24, color = 'r')
plt.xticks(index, label, fontsize=15)
plt.xlabel('선수명', fontsize = 18)
plt.ylabel('득점수', fontsize = 18)
plt.show()

```

```python
# EPL 프리미어리그 도움 순위 Top5
data_grp = df.groupby(['선수명']).sum()
assist_data_srt = data_grp.sort_values(by='도움', ascending=False).head(5)
assist_data_srt.drop(['경기수', '득점', '공격포인트', '슈팅', '유효슈팅'], axis=1, inplace=True)

assist_data_srt
```

```python
# EPL 프리미어 리그 도움 순위 Top5 막대 그래프
label = ['케인', '김덕배', '페르난데스', '그릴리쉬', '손흥민']

index = np.arange((len(label)))

plt.rcParams["font.family"] = 'AppleGothic'
plt.bar(index, assist_data_srt['도움'])
plt.title('EPL 도움 순위 Top5', fontsize = 24, color = 'r')
plt.xticks(index, label, fontsize=15)
plt.xlabel('선수명', fontsize = 18)
plt.ylabel('도움수', fontsize = 18)
plt.show()
```

```python
shot_info = df['슈팅'].describe()
shot_info
```

```python
# EPL 프리미어 리그 도움 순위 Top5 막대 그래프
label = ['총계', '평균', '표준편차', '최솟값', '25%', '50%', '75%', '최댓값']

index = np.arange((len(label)))

plt.rcParams["font.family"] = 'AppleGothic'
plt.bar(index, shot_info['슈팅'])
plt.title('EPL 슈팅률 분석', fontsize = 24, color = 'r')
plt.xticks(index, label, fontsize=15)
plt.xlabel('선수명', fontsize = 18)
plt.ylabel('도움수', fontsize = 18)
plt.show()
```
