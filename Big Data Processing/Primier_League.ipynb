```python
import requests
from bs4 import BeautifulSoup
import csv

list_data=[]
data=[]

url = "https://www3.kumhotire.com/ko/brand/sports/tottenham-player.do?seasonKey=20202021&leagueCd=EPL"

res = requests.get(url)
res.raise_for_status()

soup = BeautifulSoup(res.text, 'html.parser')

playerLists = soup.find_all('tbody', {"id":"playerList"})
playerLists
```

```python
tacs = soup.find_all('tr')
tacs

for tac in tacs:
    name = tac.find("span", {"class" : "txt_name"})
    match_nums = tac.find('td', {"class":"gp"})
    goals = tac.find('td', {"class":"gf active"})
    assists= tac.find('td', {"class":"ast"})
    attack_points = tac.find('td', {"class":"opts"})
    shootings = tac.find('td', {"class":"sht"})
    effective_shootings = tac.find('td', {"class":"sog"})

    if name != None:
        list_data = [name.get_text(), match_nums.get_text(), goals.get_text(), assists.get_text(), attack_points.get_text(), shootings.get_text(), effective_shootings.get_text()]
        data.append(list_data)

data = data[100:200]
data
```

```python
print("EPL 득점 순위")
print("선수명\t\t\t\t경기수\t득점\t\t도움\t\t공격포인트\t\t슈팅\t\t유효슈팅")
print("-"* 70)
for i,j,k,l,m,n,o, in data:
    print("{:<15}\t{}\t\t{}\t\t{}\t\t{}\t\t\t{}\t\t{}".format(i,j,k,l,m,n,o))
```
